Pre- Processing

**1. Remove stopword**

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    tokens = word_tokenize(text)
    return " ".join(tokens)
```

**2. Apply TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

def tfidf_summarize(text, top_n=2):
    sentences = sent_tokenize(text)
    cleaned = [preprocess(s) for s in sentences]

    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(cleaned)
    scores = tfidf.sum(axis=1).A1

    idx = np.argsort(scores)[-top_n:]
    idx = sorted(idx)

    return " ".join([sentences[i] for i in idx])
```

**3. Apply TextRank**

```python
!pip install summa
from summa.summarizer import summarize

def textrank_summarize(text, ratio=0.3):
    return summarize(text, ratio=ratio)
```

**4, Apply LSA Summarization**

```python
!pip install sumy
from sumy.summarizers.lsa import LsaSummarizer
from sumy.nlp.tokenizers import Tokenizer
from sumy.parsers.plaintext import PlaintextParser

def lsa_summarize(text, sentences_count=2):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, sentences_count)
    return " ".join(str(s) for s in summary)
```

**5. Apply LexRank Summarization (sumy)**

```python
from sumy.summarizers.lex_rank import LexRankSummarizer

def lexrank_summarize(text, sentences_count=2):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LexRankSummarizer()
    summary = summarizer(parser.document, sentences_count)
    return " ".join(str(s) for s in summary)
```

In this assignment, four extractive summarization algorithms were applied: TF-IDF, TextRank, LSA, and LexRank. Before performing summarization, all documents were preprocessed by converting text to lowercase, removing non-alphabetic characters, and tokenizing sentences and words. This preprocessing step ensures that the algorithms focus on meaningful lexical information rather than noise.

The TF-IDF summarization method selects sentences containing high-weighted words that frequently appear in a document but are less common across sentences. This makes TF-IDF simple and fast, but also somewhat rigid. It tends to extract sentences with repeated terms rather than sentences that best represent the overall context. In the long texts (text1 and text4), TF-IDF produced summaries that were narrow in scope and focused mostly on sentences containing technical keywords.

The TextRank algorithm, which builds a similarity graph between sentences, produced the most coherent summaries. Because it considers sentence-to-sentence relationships, TextRank is better at preserving the global structure of the document. For the longest texts, its summaries presented a logical flow that captured the main ideas more effectively than TF-IDF.

The LSA method, which uses latent semantic analysis via singular value decomposition (SVD), produced topic-oriented summaries. LSA emphasizes underlying semantic structure rather than sentence-level importance. As a result, LSA occasionally generated summaries that were short or overly focused on specific themes, particularly in longer documents. However, for shorter documents (text2 and text3), its performance remained stable.

The LexRank approach, similar to TextRank but based on TF-IDF cosine similarity, produced balanced and concise summaries. It captured central sentences effectively but tended to select fewer sentences than TextRank. In many cases, LexRank favored sentences that reflected the main thematic ideas without redundancy.

Overall, TextRank provided the best general summarization quality, capturing both sentence-level relevance and document structure. LSA was best for topic-oriented extraction, TF-IDF was simplest but most limited, and LexRank was stable and concise. Comparing them highlights how algorithmic design influences the type of summary produced.

**1. Define Positive, Negative Lexicon**

```python
positive_lexicon = {
    "love": 2, "amazing": 3, "great": 3, "helpful": 2,
    "happy": 2, "fantastic": 3, "good": 1, "transforming": 2
}

negative_lexicon = {
    "hate": -3, "terrible": -3, "bad": -2, "concerns": -1,
    "displacement": -2, "ethical": -1
}
```

**2. Apply Lexicon Sentiment**

```python
def lexicon_sentiment(text):
    tokens = preprocess(text).split()
    score = 0
    for t in tokens:
        if t in positive_lexicon:
            score += positive_lexicon[t]
        elif t in negative_lexicon:
            score += negative_lexicon[t]

    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"
```

**1. Apply BERT**

```python
!pip install transformers torch --quiet
from transformers import pipeline
bert_pipe = pipeline("sentiment-analysis")
```

**1-1. Do Summarization By TextRank & Make BERT Sentiment Function**

```python
def bert_sentiment_after_summary(text, ratio=0.2):
    # 1) Summarization (TextRank)
    summary = textrank_summarize(text, ratio=ratio)

    # Summarization Í≤∞Í≥ºÍ∞Ä ÎÑàÎ¨¥ ÏßßÏúºÎ©¥ fallback
    if summary is None or len(summary.strip()) < 5:
        summary = text[:500]

    # 2) BERT Sentiment Analysis
    result = bert_pipe(summary)[0]
    return {
        "summary": summary,
        "bert_label": result["label"],
        "bert_score": result["score"]
    }
```

**1-3. Show Conclusion**

```python
print("\n" + "="*120)
print("üìå FINAL ‚Äî Summarization ‚Üí BERT Sentiment Analysis")
print("="*120)

for name, txt in texts.items():
    print(f"\n\n========== {name.upper()} ==========\n")

    out = bert_sentiment_after_summary(txt, ratio=0.2)

    print("‚ñ∂ SUMMARY USED FOR BERT:\n")
    print(out["summary"])
    print("\n‚ñ∂ BERT RESULT:")
    print(f"   Label: {out['bert_label']}")
    print(f"   Score: {out['bert_score']:.4f}")

    print("="*80)
```

**2. Apply VADER**

```python
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()

def vader_sentiment(text):
    score = vader.polarity_scores(text)
    return score["compound"]
```

**3. Apply TextBlob**

```python
!pip install textblob
from textblob import TextBlob

def textblob_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

**4, Compare All Sentiment Analysis**

```python
# ============================================================
# 1. Define Conversion Functions (Score ‚Üí Label)
# ============================================================

def vader_to_label(score):
    if score > 0.05:
        return "positive"
    elif score < -0.05:
        return "negative"
    else:
        return "neutral"

def textblob_to_label(score):
    if score > 0:
        return "positive"
    elif score < 0:
        return "negative"
    else:
        return "neutral"


# ============================================================
# 2. Make Final Combined Sentiment Table
#    (Already have lexicon_sentiment / vader_sentiment / textblob_sentiment / BERT)
# ============================================================

rows = []
for name, txt in texts.items():
    # Lexicon-based
    lexicon_label = lexicon_sentiment(txt)

    # VADER
    vader_score = vader_sentiment(txt)

    # TextBlob
    textblob_score = textblob_sentiment(txt)

    # BERT summary ‚Üí sentiment
    bert_out = bert_sentiment_after_summary(txt, ratio=0.2)
    bert_label = bert_out["bert_label"]
    bert_score = bert_out["bert_score"]

    rows.append({
        "Text": name,
        "Lexicon": lexicon_label,
        "VADER": vader_score,
        "TextBlob": textblob_score,
        "BERT_Label": bert_label,
        "BERT_Score": bert_score
    })

all_sentiment_table = pd.DataFrame(rows)


# ============================================================
# 3. Convert Scores ‚Üí Labels for all models
# ============================================================

all_sentiment_table["VADER_Label"] = all_sentiment_table["VADER"].apply(vader_to_label)
all_sentiment_table["TextBlob_Label"] = all_sentiment_table["TextBlob"].apply(textblob_to_label)

# (BERT_Label already exists)


# ============================================================
# 4. Show Final Tables
# ============================================================

print("\n\n===================== üìå Sentiment Score Table =====================\n")
display(all_sentiment_table)

sentiment_label_table = all_sentiment_table[[
    "Text",
    "Lexicon",
    "VADER_Label",
    "TextBlob_Label",
    "BERT_Label"
]]

print("\n\n===================== üìå Sentiment Label Comparison =====================\n")
display(sentiment_label_table)
```

This aligns with known limitations of VADER when used outside social media contexts.

TextBlob uses a lexicon-based polarity model similar to VADER but produces smoother and more conservative polarity values. Its scores in this assignment ranged from mildly negative to moderately positive. TextBlob behaved more reasonably than VADER yet still could not interpret deeper contextual meaning.

In contrast, the BERT sentiment model, applied after summarization to avoid the 512-token sequence limit, provided the most contextually accurate results. BERT classified text1 and text2 as negative, reflecting the presence of problem statements, risks, and concerns across the summarized content. Meanwhile, text3 and text4 were recognized as positive, capturing their encouraging and emotionally supportive themes. Unlike the other methods, BERT considers semantic meaning and sentence-level context, leading to more nuanced and realistic conclusions.

The results clearly show the distinction between lexicon-based methods and contextual models:

Lexicon, VADER, and TextBlob rely on word lists and produce surface-level sentiment.

BERT interprets the true meaning of the text.

This comparison demonstrates how modern NLP techniques significantly outperform traditional approaches, particularly for long and complex documents.
